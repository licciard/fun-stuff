```
import mpmath 
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
import math
import scipy.special as scp

%matplotlib inline
```

Polynomials

```
def B0(x):
  return 1

def B1(x):
  return x - 1/2

def B2(x):
  return mpmath.power(x,2)-x+1/6

def B3(x):
  return mpmath.power(x,3)-3/2*mpmath.power(x,2) +1/2*x

def B4(x):
  return mpmath.power(x,4)-2*mpmath.power(x,3)+mpmath.power(x,2)-1/30


xmin = -10
xmax = 10

dx = np.arange(xmin,xmax,0.01)
b1 = np.array([B1(x-mpmath.floor(x)) for x in dx])
b2 = np.array([B2(x-mpmath.floor(x)) for x in dx])
b3 = np.array([B3(x-mpmath.floor(x)) for x in dx])
b4 = np.array([B4(x-mpmath.floor(x)) for x in dx])

plt.figure(figsize=(15,10))
#plt.plot(dx, b1, 'g-')
#plt.plot(dx, b2, 'b-')
plt.plot(dx, b3, 'k-')
plt.plot(dx, b4, 'r-')

#plt.xscale('log')
#plt.yscale('log')

axes = plt.gca()
#axes.set_xlim([0,10])
#axes.set_ylim([0,5])
plt.show()
```
