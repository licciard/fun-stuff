```
mpmath.mp.dps = 15

# for zeta tildes
def ZtM(i,j):
  if i == j:
    if i == 0: return 1
  if i < j: return 0
  if i > 0 and j == 0: return 0
  return Zt[i-j]

# for pascal triangle
def PtM(i,j):
  return sympy.binomial(i,j)

def PtMI(i,j):
  return sympy.binomial(i,j)*(1,-1)[(j)%2]

```

```
Nmax = 20
Zm = sympy.Matrix(Nmax,Nmax,ZtM)
Pm = sympy.Matrix(Nmax,Nmax,PtM)
PmI = sympy.Matrix(Nmax,Nmax,PtMI)
ZmI = Zm.inv()

lv = sympy.Matrix([ls[i] for i in range(Nmax)])

ZmI
```

```
mpmath.mp.dps = 25


def inv_mat_Zt(n):
  if n == 1:
    return -2
  if n == 2:
    return 2*mpmath.log(2*mpmath.pi)
  tot = 0
  for k in range(1,n):
    tot += (Zt[k-1]-Zt[k])*ZmIs[n-k] #inv_mat_Zt(n-k)
  ZmIs.append(-2*tot)
  return ZmIs[-1]

def inv_zeta(x):
  return 1/mpmath.zeta(x)

ZmIs = [0,-2,2*mpmath.log(2*mpmath.pi)]
It = []
for n in range(1,20):
  It.append(inv_mat_Zt(n))
  print("n:{} --> {} vs {}".format(n,It[-1],mpmath.diff(inv_zeta,0,n=n-1)/mpmath.gamma(n)))
```

```
# comparison of exponential terms
mpmath.mp.dps = 25



def exp_zeta(n):
  tot = 0
  for p in range(2,n+1):
    tot += sympy.binomial(n,p)*mpmath.zeta(p)*mpmath.power(1/2,p-1) #(1-mpmath.power(1/2,p-1))*(1,-1)[p%2]
  return tot

def exp_deriv(n):
  tot = 0
  for k in range(n):
    tot += sympy.binomial(n-1,k)*(k+1)*It[k+1]*(1,-1)[k%2]
  return tot

for n in range(2,200):
  print("n: {} --> {}".format(n,exp_zeta(n) - exp_deriv(n)))
  #print("n: {} --> {} vs {}".format(n,exp_zeta(n)/mpmath.power(3/2,n)/2,exp_deriv(n)/mpmath.power(3/2,n)))
```

```
def my_eta(n):
  val = 0
  for k in range(n+2):
    val -= k*It[k]*Zt[n+1-k]
  for k in range(n+1):
    term = 0
    for m in range(1,n-k+1):
      term += It[k]*Zt[n-k-m]*mpmath.zeta(1+m)*(1-mpmath.power(1/2,m+1)*(1+(-1,1)[m%2]))
    val -= term
  return val

def my_eta_cancel(n):
  val = 0
  for k in range(n+2):
    val -= k*It[k]*Zt[n+1-k]
  if n > 0:
    val -= mpmath.zeta(1+n)*(1-mpmath.power(1/2,n+1)*(1+(-1,1)[n%2]))
  else: 
    val -= (mpmath.euler + mpmath.log(mpmath.pi*2))
  return val*(1,-1)[n%2]

def my_eta_exp(n):
  val = 0
  for k in range(n+2):
    val += k*It[k]*Zt[n+1-k]
  return val

for n in range(50):
  print("n: {} --> {} vs {} & {}".format(n,my_eta(n),my_eta_cancel(n),my_eta_exp(n)))
```
