Zeta function

```
mpmath.mp.dps = 20

s = 2
def zeta_ap_func(t):
  sine = mpmath.sin(s*mpmath.atan(t))
  deno = mpmath.power(1+t*t,s/2)
  apexp = mpmath.exp(2*mpmath.pi*t)-1
  return (sine/deno) / apexp

def zeta_ap_split_zero(t):
  term = mpmath.exp(mpmath.j*mpmath.atan(t))/mpmath.sqrt(1+t*t)
  apexp = mpmath.exp(2*mpmath.pi*t)-1
  return mpmath.power(term,s)/apexp

def zeta_ap_split_inf(t):
  term = mpmath.exp(mpmath.j*mpmath.atan(1/t))*t/mpmath.sqrt(1+t*t)
  apexp = mpmath.power(t,2)*(mpmath.exp(2*mpmath.pi/t)-1)
  return mpmath.power(term,s)/apexp

def zeta_ap(z):
  term = 1./(z-1)
  const = 1./2
  #apterm = mpmath.quad(zeta_ap_func,[0,1e5])
  #print(2*apterm)
  apterm_zero = mpmath.im(mpmath.quad(zeta_ap_split_zero,[0,1]))
  apterm_inf = mpmath.im(mpmath.quad(zeta_ap_split_inf,[0,1]))
  #print("{} vs {}".format(apterm_zero,apterm_inf))
  return term+const+2*(apterm_zero+apterm_inf)

for n in range(2,25):
  s = n  
  print("n: {} --> {} vs {}".format(n,zeta_ap(n),mpmath.zeta(n)))
```
