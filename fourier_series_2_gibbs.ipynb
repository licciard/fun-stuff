```
import numpy as np
import matplotlib.pyplot as plt
```

```
xmin = -10
xmax = 10
npoints = 50000

x = np.linspace(xmin, xmax, npoints)

def f(xi):
  xp = (xi+np.pi)//(2*np.pi)
  return xi - xp*2*np.pi

def fourier(xi,Nmax = 5):
  a_0 = 0
  tot = a_0/2
  for n in range(1,Nmax+1):
    a_n = 0
    b_n = -2/n * (-1)**n
    tot += a_n * np.cos(n*xi) + b_n * np.sin(n*xi)
  return tot

plt.figure(figsize=(10,5))
plt.plot(x,f(x),c='b')
plt.plot(x,fourier(x,2),c='r')
plt.plot(x,fourier(x,25),c='g')
plt.xlabel('x'); plt.ylabel('y');
plt.ylim([-4.25,4.25])
plt.show()
```

```
ns = np.arange(1,200)
diff = []
for n in ns:
  yf  = fourier(x,n)
  diff.append((max(yf)-np.pi)/(2*np.pi))
```

```
plt.figure(figsize=(10,10))
plt.plot(ns,diff,c='b')
plt.ylim([0.08,0.09])
plt.show()
```
